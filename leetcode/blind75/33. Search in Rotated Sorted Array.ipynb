"* nlogn solution required.
* Took 3hrs to handle edge cases hence marked as hard.

* Soltion is straight forward
1. [important] Use a type of Binary Search to find the pivot index.  -> left pointer will be the pivot at the end
2. Figure out in which region our target element will be.
3. Perform vanilla Binary Search on that region."

```python
from typing import *

class Solution:
    def search(self, nums: List[int], target: int) -> int:
        
        # ** Double Binary Search Solution (Approach 1) **
        
        # 1. Find the Pivot Index using Binary Search
        i_piv = self.pivot_index(nums)
        
        
        # 2. Figure out which side of the pivot our Target Element is.
        i_start = 0
        i_end = len(nums) - 1
        
        
        if ( (nums[0] <= target) and (target <= nums[i_piv - 1]) and (i_piv > 0)):
            # (i_piv > 0) -> to handle edge case when len(nums) = 1 
            i_start = 0
            i_end = i_piv - 1
        
        elif( (nums[i_piv] <= target) and (target <= nums[-1]) and (i_piv > 0)):
            i_start = i_piv
            i_end = len(nums) - 1
            
         
        
        # 3. Perform Binary Search on the correct side
        ans = self.binary_search(nums, i_start, i_end, target)
        
        
        # 4. Return the ans
        return ans
    
    
    
    def pivot_index(self, nums):
        # Dry run this to understand the code
        l = 0
        r = len(nums) - 1
        p = 0
        
        while(l < r):
            p = (l+r)//2
            
            l_num = nums[l]
            p_num = nums[p]
            r_num = nums[r]
            
            # p_num > r_num when pivot is in the left region
            # so we need to go to the right region
            if(p_num > r_num):
                l = p + 1
            
            else:
                r = p
        
        # At the end left_index will be the pivot
        return l
    
    
    
    
    def binary_search(self, nums, left, right, target):
        
        while(left <= right):
            mid = (left + right)//2
            
            l_num = nums[left]
            m_num = nums[mid]
            r_num = nums[right]
            
            if(m_num == target):
                return mid
            
            elif(m_num > target):
                right = mid - 1
            
            else:
                left = mid + 1
            
        
        return -1
```
