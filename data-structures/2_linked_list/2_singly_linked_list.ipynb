```python
class Node:
    # Constructor
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class SinglyLinkedList:
    
    # Constructor
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    
    
    # Insert
    def insert(self, data):
        newNode = Node(data)
        
        if self.head == None:
            self.head = newNode
            self.tail = newNode
        
        else:
            self.tail.next = newNode
            self.tail = newNode
        
        self.length += 1
    
    
    
    # Print LL
    def printLL(self):
        
        currNode = self.head
        
        while(currNode != None):
            print(currNode.data, end="->")
            currNode = currNode.next
            
        print("NULL")
    
    
    
    # Input a LL
    def inputLL(self):
        
        # Take LL as input and store it in a List
        inputList = [int(num) for num in input().split()]   # List Comprehension
        
        for num in inputList:
            self.insert(num)
        
        return self.head
    
    
    
    
    # Insert at ith Position (Iterative Approach)
    def insert_at_ith(self, arg_data, arg_pos):
        
        if(arg_pos <= 0):
            newNode = Node(arg_data)
            newNode.next = self.head
            self.head = newNode
            return
        
        currNode = self.head
        
        for i in range(0, arg_pos-1):
            currNode = currNode.next
            
            if (i >= self.length):
                # If given position is beyond the length of LL
                # Insert the element at the end
                break
            
             
        nextNode = currNode.next
        
        newNode = Node(arg_data)
        
        currNode.next = newNode
        newNode.next = nextNode
        
    
    
    
    # Insert at ith Position (Recursive Approach)
    def helper_insert_at_ith_recursive(self, arg_head, arg_data, arg_pos):
        
        # 1. Base Case - insert the new node here
        if(arg_pos == 0):
            # Insert the newNode before current head
            newNode = Node(arg_data)
            newNode.next = arg_head
            
            return newNode # The new head of the LL
        
        
        # Edge Case - So that we dont do arg_head.next on None
        if(arg_head is None):
            return None
        
        
        # 2. Recursive Case - recursion will insert the new node when it hits the base case
        sub_ll_head = self.helper_insert_at_ith_recursive(arg_head.next, arg_data, arg_pos - 1)
        
        
        # 3. Current Problem - Connect head to the sub-ll in which the new node is inserted
        arg_head.next = sub_ll_head
        
        
        return arg_head
    
    
    
    def insert_at_ith_recursive(self, arg_data, arg_pos):
        
        if(arg_pos == 0):
            self.head = self.helper_insert_at_ith_recursive(self.head, arg_data, arg_pos)
        else:
            self.helper_insert_at_ith_recursive(self.head, arg_data, arg_pos)
        
        
    
            
            
            
    
    # Delete the ith Node (Iterative Approach)
    def delete_at_ith(self, arg_pos):
        
        # Edge Case
        if(arg_pos == 0):
            delNode = self.head
            self.head = self.head.next
            
            del delNode
            return
        
        
        prevNode = self.head
        currNode = self.head
        nextNode = self.head.next
        curr_pos = 0
        
        while (curr_pos < arg_pos):
            prevNode = currNode
            
            currNode = currNode.next
            
            if(currNode == None):
                nextNode = None
            else:
                nextNode = currNode.next
                
            
            curr_pos += 1
        
        # Delete currNode
        prevNode.next = nextNode
        del currNode
        
        return
    
    
    
    
    # Delete the ith Node (Recursive Approach)
    def helper_delete_at_ith_recursive(self, arg_head, arg_pos):
        
        # 1. Base Case
        if(arg_pos == 0):
            arg_head = arg_head.next
            return arg_head
        
        
        # 2. Recursive Case
        sub_ll_head = self.helper_delete_at_ith_recursive(arg_head.next, arg_pos - 1)
        
        
        # 3. Current Problem
        arg_head.next = sub_ll_head
        
        return arg_head
        
    
    
    def delete_at_ith_recursive(self, arg_pos):
        
        if(arg_pos == 0):
            self.head = self.helper_delete_at_ith_recursive(self.head, arg_pos)
        
        else:
            self.helper_delete_at_ith_recursive(self.head, arg_pos)
        
    

```

```python
def main():
    ll = SinglyLinkedList()
    ll2 = SinglyLinkedList()
    
    ll.insert(10)
    ll.insert(20)
    ll.insert(30)
    ll.insert(40)
    
    ll.printLL()
    
    ll.insert_at_ith(22, 2)
    ll.printLL()
    ll.insert_at_ith(9, 0)
    ll.printLL()
    ll.insert_at_ith(59, 10)
    ll.printLL()
    
    ll.insert_at_ith_recursive(87, 0)
    ll.printLL()
    
    ll.insert_at_ith_recursive(88, 1)
    ll.printLL()
    
    ll.insert_at_ith_recursive(89, 2)
    ll.printLL()
    
    ll.delete_at_ith(9)
    ll.printLL()
    
    ll.delete_at_ith(0)
    ll.printLL()
    
    ll.delete_at_ith_recursive(0)
    ll.printLL()
    
    ll.delete_at_ith_recursive(1)
    ll.printLL()
    
    ll.delete_at_ith_recursive(5)
    ll.printLL()
    
#     ll2.inputLL()
#     ll2.printLL()

if __name__=="__main__":
    main()
```
